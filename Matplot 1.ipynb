# Generate 50 random X and Y values between 0 to 50.Set marker tranparent(alpha) to 0.5

# Label Each point with its(x,y) coordinates using annotation 
# x= [5,10,15,20,25], y = [50,100,150,200,250].

```python
x=[5,10,15,20,25]
y=[50,100,150,200,250]
plt.scatter(x,y,color='red',marker='o')
for i in range(len(x)):
    plt.annotate(f'({x[i]},{y[i]})',(x[i],y[i]),ha='center')
plt.title('Scatter Plot with Annotation')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid()
plt.show()
```

# Create a scatter plot for:
# X=[2,4,6],Y=[10,20,30].
# Add text labels['A','B','C'] for each point.

```python
x=(2,4,6)
y=(10,20,30)
l=['A','B','C']
plt.scatter(x,y,color='red',marker='*')
for i in range(len(x)):
    plt.annotate(f'({x[i]},{y[i]},{l[i]})',(x[i],y[i],),ha='center')
plt.title('Scatter plot with Label')    
plt.grid()
plt.show()     
```

# The Average temperature in a city for the six months of the year is as follows:

```python
data={
    'Month':['January','February','March','April','May','June'],
    'Temperature(C)':[5,7,10,15,20,25]
}
df=pd.DataFrame(data)
df
    
```

```python
new=['pink']*5+['orange']
plt.bar(df['Month'],df['Temperature(C)'],color=new)
plt.title('Temperature in city')
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.show()
```

```python
Month =['January','Februrary','March','April','May','June']
Temperature =[5,7,10,15,20,25]
new=[]
for i in Temperature:
    if i>20:
        new.append('red')
    else:
        new.append('blue')
plt.bar(Month,Temperature,color=new)
plt.title('Temperature of each month')
plt.xlabel('Months')
plt.ylabel('Temperature (C)')
plt.show()      
```

```python
data={
    'Company':['Company A','Company B','Company C','Company D','Company E'],
      'Profit':[12,8,15,10,9]
}
df=pd.DataFrame(data)
df
```

# Create a horizontal bar graph to repersent this data.
# sort the bars in decs order of profit before plotting.

```python
df_sorted=df.sort_values(by='Profit',ascending=False)
df_sorted
```

```python
plt.barh(df_sorted['Company'],df_sorted['Profit'],color='skyblue')
plt.title('Companies Profit in million USD')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()
```

# Create pie chart to represent this data.
# Highlight the segment for Brand A.
# Add labels and percentage to the chart.
# Use a suitable title for the chart.

```python
y=np.array([35,25,20,15,5])
mylabels=['Brand A','Brand B','Brand C','Brand D','Other Brand']
myexplode=[0.2,0,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode,shadow=True,autopct='%1.2f%%')
plt.legend(title="Market Share of different smartphones brands",loc='upper left',bbox_to_anchor=(1,1))
plt.show()
```

# Q.1. Monthly Expenditure Breakdown
# A person spends their monthly income on the following categories:
# Rent: 40%
# Food: 25%
# Transportation: 15%
# Entertainment: 10%
# Savings: 10%
# Write a python script to:

```python
Categories=['Rent','Food','Transportation','Entertainment','Savings']
x=[40,25,15,10,10]
plt.pie(x,labels=Categories,shadow=True,autopct='%1.2f%%')
plt.legend(title="Monthly Expenditure Breakdown",loc='upper left',bbox_to_anchor=(1,1))
plt.show()
```

# Histogram

```python
ages=[12,18,22,25,28,30,35,40,45,50]
bins=[10,20,30,40,50,60] #10-12,18 ,20-22,25,28  ,30-30,35 so on
plt.hist(ages,bins=bins,edgecolor='black',color='red')
plt.title('Histtogram of Ages')
plt.xlabel('Age Range (Bins)')
plt.ylabel('Frequency')
plt.show()
```
